```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train[10]
```

```python
x_train[11]
```

```python
x_train[3]
```

```python
x_train = tf.keras.utils.normalize(x_train, axis = 1)

x_test = tf.keras.utils.normalize(x_test, axis = 1)
```

```python
# create the model architecture

DL_model1 = tf.keras.models.Sequential()
DL_model1.add(tf.keras.layers.Flatten())
DL_model1.add(tf.keras.layers.Dense(128, activation = tf.nn.relu))
DL_model1.add(tf.keras.layers.Dense(64, activation = tf.nn.relu))
DL_model1.add(tf.keras.layers.Dense(10, activation = tf.nn.softmax))
```

```python
DL_model1.compile( loss = "sparse_categorical_crossentropy", optimizer = "adam", metrics = "accuracy")
DL_model1.fit(  x_train, y_train, epochs = 20, batch_size = 1000   )
```

```python
plt.imshow(x_train[10])
```

```python
DL_model1.evaluate(x_train, y_train), DL_model1.evaluate(x_test, y_test)
```
